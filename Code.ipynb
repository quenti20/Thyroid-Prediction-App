```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
df['Sex'] = label_encoder.fit_transform(df['Sex'])
df['Thyroid'] = label_encoder.fit_transform(df['Thyroid'])
```

```python
import pandas as pd
from scipy.stats import f_oneway, chi2_contingency

# ANOVA test for continuous features
continuous_features = ['Age', 'Height', 'Weight', 'BMI', 'Body fat', 'Serum T3', 'Serum T4', 'TSH']
anova_results = []
for feature in continuous_features:
    hypothyroid = df[df['Thyroid'] == 1 ][feature]
    normal = df[df['Thyroid'] == 0 ][feature]
    if len(set(hypothyroid)) > 1 and len(set(normal)) > 1:  # Check for variability
        f_statistic, p_value = f_oneway(hypothyroid, normal)
        anova_results.append({'Feature': feature, 'F-Score': f_statistic, 'p-value': p_value})
    else:
        anova_results.append({'Feature': feature, 'F-Score': 'Insufficient Variability', 'p-value': 'Insufficient Variability'})

anova_df = pd.DataFrame(anova_results)
print("ANOVA Test Results:")
print(anova_df)
```

```python
# Chi-square test for categorical features
chi2_results = {}
categorical_features = ['Sex']
for feature in categorical_features:
    contingency_table = pd.crosstab(df[feature], df['Thyroid'])
    chi2, p, _, _ = chi2_contingency(contingency_table)
    chi2_results[feature] = {'Chi2 value': chi2, 'P-value': p, 'Degrees of freedom': contingency_table.shape[1] - 1}
print("\nChi-square Test Results:")
for feature, result in chi2_results.items():
    print("Feature:", feature)
    print("Chi2 value:", result['Chi2 value'])
    print("P-value:", result['P-value'])
    print("Degrees of freedom:", result['Degrees of freedom'])
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import f_classif

# Assuming df is already loaded and encoded as per your context

# Selecting numeric features
numeric_features = df.select_dtypes(include=[np.number])

# Performing ANOVA test
f_scores, p_values = f_classif(numeric_features, df['Thyroid'])

# Creating a DataFrame to display results
anova_results = pd.DataFrame({
    'Feature': numeric_features.columns,
    'F-Score': f_scores,
    'p-value': p_values
})

# Displaying results
print("ANOVA Test Results:")
print(anova_results)
```

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Assuming df is already loaded and encoded as per your context

# Contingency table
contingency_table = pd.crosstab(df['Thyroid'], df['Sex'])

# Performing chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

# Displaying results
print("Chi-square Test Results:")
print(f"Chi2 value: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:")
print(expected)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Separate features (X) and target variable (Y)
X = df.drop('Thyroid', axis=1)
Y = df['Thyroid']

# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Initialize models
models = {
    'Random Forest': RandomForestClassifier(random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
    'AdaBoost': AdaBoostClassifier(random_state=42),
    'Logistic Regression': LogisticRegression(random_state=42),
    'XGBoost': XGBClassifier(random_state=42),
    'LightGBM': LGBMClassifier(random_state=42)
}

# Train and evaluate models
results = {}
for name, model in models.items():
    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_test)
    accuracy = accuracy_score(Y_test, Y_pred)
    precision = precision_score(Y_test, Y_pred, pos_label=1, average='binary')
    recall = recall_score(Y_test, Y_pred, pos_label=1, average='binary')
    f1 = f1_score(Y_test, Y_pred, pos_label=1, average='binary')
    results[name] = {'Accuracy': accuracy, 'Precision': precision, 'Recall': recall, 'F1-score': f1}

# Display results
results_df = pd.DataFrame(results)
print(results_df)
```

```python
import numpy as np

# Setting up the positions for the bars
bar_width = 0.3
index = np.arange(len(models))

# Plotting the bar graph
plt.figure(figsize=(18, 8))

# Plotting accuracy
bars_accuracy = plt.bar(index - bar_width, results_df.loc['Accuracy'], bar_width, label='Accuracy', color='skyblue')

# Plotting precision
bars_precision = plt.bar(index, results_df.loc['Precision'], bar_width, label='Precision', color='salmon')

# Plotting recall
bars_recall = plt.bar(index + bar_width, results_df.loc['Recall'], bar_width, label='Recall', color='lightgreen')

# Adding labels and ticks
plt.xlabel('Models')
plt.ylabel('Scores')
plt.title('Performance Comparison of Different Models')
plt.xticks(index, models)
plt.legend()

# Function to add text annotations
def add_labels(bars):
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 3), ha='center', va='bottom')

# Adding precision, recall, and F1-score annotations
add_labels(bars_accuracy)
add_labels(bars_precision)
add_labels(bars_recall)

plt.show()
```
